```python
x = data.iloc[:,[2,3]].values
```

```python
x
```

```python
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
sns.set()

plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)

y = kmeans.fit_predict(x)

y
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0], x[y==0,1], s=50, c='green', label='First Cluster')
plt.scatter(x[y==1,0], x[y==1,1], s=50, c='red', label='Second Cluster')
plt.scatter(x[y==2,0], x[y==2,1], s=50, c='yellow', label='Third Cluster')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='centroid')

plt.title('Flowers type')
plt.xlabel('Iris')
plt.ylabel('Flowers')
plt.show()
```
