```python
x_train,x_test,y_train,y_test = train_test_split(x1,y1,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
x1_prediction = model.predict(x_train)
```

```python
x1_accuracy = accuracy_score(x1_prediction,y_train)
```
